### Creaci√≥n del modelo

#### Algoritmos disponibles

```python
models()
```

#### Latent Dirichlet Allocation (lda)

```python
lda = create_model("lda", num_topics = 5, multi_core = True)
```

```python
print(lda)
```

### Asignaci√≥n del modelo

```python
lda_resultados = assign_model(lda)
lda_resultados.head()
```
